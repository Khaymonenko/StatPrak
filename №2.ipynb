```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch
def beauty_derivative_plot(x_point):
    x = np.linspace(-2 * np.pi, 2 * np.pi, 400)
    y = 2 * np.sin(x) - np.cos(2 * x)
    dy = 2 * np.cos(x) + 2 * np.sin(2 * x)
    fx = 2 * np.sin(x_point) - np.cos(2 * x_point)
    x_tangent = np.linspace(x_point - 1, x_point + 1, 100)
    y_tangent = (fx) + (2 * np.cos(x_point) + 2 * np.sin(2 * x_point))  * (x_tangent - x_point)
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, color='blue', label='f(x) = 2sinx - cos2x')
    plt.plot(x, dy, linestyle='dashed', color = 'red', label="f'(x) = 2cos x + 2sin 2x")
    plt.plot(x_tangent, y_tangent, 'g', label=f'Tangent at x = {x_point:.2f}')
    plt.scatter(x_point, fx, color='black')
    plt.plot([x_point, x_point], [0, fx], linestyle='dashed', color = 'purple', linewidth=1.5)
    plt.plot([0, x_point], [fx, fx], linestyle='dashed', color = 'purple', linewidth=1.5)
    arrow = FancyArrowPatch(
        (x_point - 0.1, fx - 1.77),
        (x_point, fx),
        connectionstyle="arc3,rad=.08",
        arrowstyle='-|>',
        color='black',
        linewidth=1.1,
        mutation_scale=10,
        zorder=10
    )
    plt.gca().add_patch(arrow)  # Добавляем стрелку на график
    plt.text(x_point - 1.2, fx - 2.10, f"f'({x_point:.2f}) = {fx:.2f}", fontsize=10)
    plt.axhline(0, color='black', linewidth=0.5) #Толстые оси
    plt.axvline(0, color='black', linewidth=0.5)
    plt.xlabel('x') # обозначения в легенде
    plt.ylabel('y')
    plt.legend(loc='upper right', fontsize='small') # легенда
    plt.title("Function, Derivative, and Tangent Line")
    plt.xlim(-2 * np.pi, 2 * np.pi)
    plt.ylim(-4, 4)
    plt.grid(True)
    plt.show()
beauty_derivative_plot(-3)
beauty_derivative_plot(-2)
beauty_derivative_plot(1.7)

```

Task 2.1
The data in populations.txt describes the populations of hares, lynxes and carrots in northern Canada during 20 years. Do the following exercises without for-loops.

Plot the population size of each species for each year on a single graph. Add legend and axis labels to the plot.

```python
data = np.loadtxt('populations.txt')
years = data[:, 0]
hares = data[:, 1]
lynxes = data[:, 2]
carrots = data[:, 3]
plt.figure(figsize=(8, 5))
plt.plot(years, hares, label='Зайчики', color='blue')
plt.plot(years, lynxes, label='Рысята', color='red')
plt.plot(years, carrots, label='Морковки', color='green')
plt.xlabel('Года')
plt.ylabel('Популяция')
plt.title('Популяция зайчиков, рысят и морковок')
plt.legend()
plt.grid(True)
plt.xlim(1900, 1920)
plt.show()
```

Task 2.2
Find which year each species had the largest population. Mark these years in the plot of populations.

```python
max_hares_year = years[np.argmax(hares)]
max_lynxes_year = years[np.argmax(lynxes)]
max_carrots_year = years[np.argmax(carrots)]
max_hares = np.max(hares)
max_lynxes = np.max(lynxes)
max_carrots = np.max(carrots)
plt.figure(figsize=(10, 5))
plt.plot(years, hares, label='Зайчики', color='blue')
plt.plot(years, lynxes, label='Рысята', color='red')
plt.plot(years, carrots, label='Морковки', color='green')
plt.scatter(max_hares_year, max_hares, color='blue', s=60, zorder=2)
plt.scatter(max_lynxes_year, max_lynxes, color='red', s=60, zorder=2)
plt.scatter(max_carrots_year, max_carrots, color='green', s=60, zorder=2)
plt.xlabel('Года')
plt.ylabel('Популяция')
plt.title('Популяция зайчиков, рысят и морковок')
plt.legend()
plt.grid(True)
plt.show()
```

Task 2.3
Assuming total population of hares and lynxes is 100%, plot stacked bar graph showing ratio of each specis for each year in the period.

```python
total = hares + lynxes
hares_ratio = hares / total
lynxes_ratio = lynxes / total
plt.figure(figsize=(10, 5))
plt.bar(years, hares_ratio, label="Зайчики", color="blue")
plt.bar(years, lynxes_ratio, bottom=hares_ratio, label="Рысята", color="red")
plt.xlabel("Года")
plt.ylabel("Доля")
plt.title("Соотношение зайчат и рысят")
plt.legend()
plt.grid(axis='y', linestyle="--", alpha=0.7)
plt.show()
```

Task 2.4
Find the mean and std of the populations of each species. Plot the historgam of population for each species and show mean values with vertical line. Show each histogram in a separate plot. Arrange 3 plots in a row.

```python
mean_hares = np.mean(hares)
mean_lynxes = np.mean(lynxes)
mean_carrots = np.mean(carrots)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].hist(hares, bins=10, color='blue', edgecolor='black')
axes[0].axvline(mean_hares, color='black', linestyle='dashed', linewidth=1, label=f'Среднее: {mean_hares:.1f}')
axes[0].set_title("Популяция зайчиков")
axes[0].set_xlabel("Численность")
axes[0].set_ylabel("Частота")
axes[0].legend()

axes[1].hist(lynxes, bins=10, color='red', edgecolor='black')
axes[1].axvline(mean_lynxes, color='black', linestyle='dashed', linewidth=1, label=f'Среднее: {mean_lynxes:.1f}')
axes[1].set_title("Популяция рысят")
axes[1].set_xlabel("Численность")
axes[1].legend()

axes[2].hist(carrots, bins=10, color='green', edgecolor='black')
axes[2].axvline(mean_carrots, color='black', linestyle='dashed', linewidth=1, label=f'Среднее: {mean_carrots:.1f}')
axes[2].set_title("Популяция морковок")
axes[2].set_xlabel("Численность")
axes[2].legend()

plt.tight_layout()
plt.show()
```

Task 2.5
Find which species (hares or lynxes) has the largest population for each year. Print the result as [H, H, L, H, ...]. Plot a pie chart showing the ratio of "H" and "L" values.

```python
result = np.where(hares > lynxes, 'H', 'L')
print(result.tolist())
unique, counts = np.unique(result, return_counts=True)
plt.figure(figsize=(6, 6))
plt.pie(counts, labels=unique, colors=['blue', 'red'])
plt.title("Соотношение лет доминирования зайчат и рысят")
plt.show()
```

Task 2.6
Plot the change in the hare population and in the lynx population for each year. Find the correlation coefficient between the both time series.

```python
hares_change = np.diff(hares)
lynxes_change = np.diff(lynxes)
years_change = years[1:]
correlation = np.corrcoef(hares_change, lynxes_change)[0, 1]
plt.figure(figsize=(10, 5))
plt.plot(years_change, hares_change, label="Изменение популяции зайчат", color="blue")
plt.plot(years_change, lynxes_change, label="Изменение популяции рысят", color="red")
plt.xlabel("Года")
plt.ylabel("Изменение популяции")
plt.title("График изменения популяций зайчат и рысят")
plt.legend()
plt.grid(True)
```

Task 2.8
Assume the population of hares in 1920 is unknown. Suggest a way to estimate this value. Compare an estimated value with the true value and print a ratio of the error to the true value.

```python
average_change = np.mean(np.diff(hares))
estimated_hares = hares[-2] + average_change
true_hares = hares[-1]
error_ratio = abs(estimated_hares - true_hares) / true_hares
print(f"Среднее изменение популяции: {average_change}")
print(f"Оценённое значение популяции зайчат в 1920 году: {estimated_hares}")
print(f"Истинное значение популяции зайчат в 1920 году: {true_hares}")
print(f"Относительная ошибка: {error_ratio}")
```
